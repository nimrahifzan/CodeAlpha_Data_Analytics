task 3 : Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
print("Graphic Libraries are loaded and ready!")
```

```python
figs, axes = plt.subplots(2, 2, figsize=(16,12))
sns.barplot(ax=axes[0,0], data=df, x='Total', y='Product line', estimator=sum, errorbar=None,palette='viridis')
axes[0,0].set_title('Total Sales Revenue by Product Line', fontsize=12, fontweight='bold')
axes[0,0].set_xlabel('Total Revenue ($)')
axes[0,0].set_ylabel('Product Category')

sns.histplot(ax=axes[0,1], data=df, x='Rating', kde=True, color='purple', bins=10)
axes[0,1].set_title('Distribution of Customer Ratings', fontsize=12, fontweight='bold')
axes[0,1].set_xlabel('Rating(1-10)')
axes[0,1].set_ylabel('Count of Transactions')

sns.countplot(ax=axes[1,0], data=df, x='Payment', hue='Branch',  palette='muted')
axes[1,0].set_title('Payment Method Prefrences across Branches', fontsize=12, fontweight='bold')
axes[1,0].set_xlabel('Payment Method')
axes[1,0].set_ylabel('Number of Transactions')

sns.scatterplot(ax=axes[1,1], data=df, x='Rating', y='gross income', hue='Customer type', alpha=0.7, palette='magma')
axes[1,1].set_title('Gross Income vs Customer Rating', fontsize=12, fontweight='bold')
axes[1,1].set_xlabel('Customer Rating')
axes[1,1].set_ylabel('Gross Income ($)')

plt.tight_layout()
plt.show()
```
